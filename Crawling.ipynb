```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time

driver = webdriver.Chrome()
url = "https://www.friendsscreen.kr/main/shop"
driver.get(url)
```

```python
info_list = []

# 첫 페이지부터 시작하여 정보 수집
page_num = 1
while (page_num <= 20 ):
    print(f"현재 페이지: {page_num}")

    # 매장 정보 요소들 찾기
    info_boxes = driver.find_elements(By.CLASS_NAME, 'info')

    # 각 매장 정보 클릭 및 정보 수집
    for info in info_boxes:
        # JavaScript를 사용하여 클릭 이벤트 실행
        driver.execute_script("arguments[0].click();", info)
        time.sleep(1)  # 페이지 로딩 대기

        # 매장 정보 가져오기
        try:
            infoBox = driver.find_element(By.CLASS_NAME, 'infoBox').text
            info_list.append(infoBox)
        except:
            info_list.append("정보를 찾을 수 없음")  # 예외 처리: 정보를 찾을 수 없을 경우 메시지 추가

    # 다음 페이지로 이동 (다음 버튼 클릭)
    try:
        next_button = driver.find_element(By.XPATH, "//a[@class='next']")
        next_button.click()
        time.sleep(3)  # 페이지 로딩 대기
        page_num += 1
    except:
        print(f"더 이상 다음 페이지가 없습니다. 총 {page_num} 페이지를 수집하였습니다.")
        break

```

```python
import pandas as pd
pd.set_option('display.max_colwidth', None) # 열 너비 제한 없음

FNAME = []
ADDRESS = []
TEL = []
OPEN_TIME = []
SENSOR = []
PARKING = []
LEFT_HAND = []
SWING_PLATE = []

for infoBox in info_list:
    item = infoBox.split('\n')
    FNAME.append(item[0])  # 첫 번째 항목은 매장명으로 가정
    ADDRESS.append(item[2])  # 세 번째 항목은 주소로 가정
    
    # 전화번호(Tel) 가져오기
    if item[3] != '전화번호':
        TEL.append('NULL')  # '전화번호'가 아닌 경우 None 처리
    else:
        TEL.append(item[4])  # '전화번호'인 경우 해당 값 가져오기
    
    # 영업시간(OPEN_TIME) 가져오기
    if item[3] != '전화번호':
        open_time = item[4:7] 
    else:
        open_time = item[6:9]  
    OPEN_TIME.append(open_time)
    
    # 센서 정보(SENSOR) 가져오기
    if item[3] != '전화번호':
        sensor = item[8:18]  
    else:
        sensor = item[10:20]  
    SENSOR.append(sensor)
    
    if item[3] != '전화번호':
        PARKING.append(item[19])  
    else:
        PARKING.append(item[21])

    if item[3] != '전화번호':
        LEFT_HAND.append(item[21])  
    else:
        LEFT_HAND.append(item[23])  
    
    if item[3] != '전화번호':
        SWING_PLATE.append(item[23])  
    else:
        SWING_PLATE.append(item[25])  

    # 데이터프레임 생성
df = pd.DataFrame({
    'FNAME': FNAME,
    'ADDRESS': ADDRESS,
    'TEL': TEL,
    'OPEN_TIME': OPEN_TIME,
    'SENSOR': SENSOR,
    'PARKING': PARKING,
    'LEFT_HAND': LEFT_HAND,
    'SWING_PLATE':SWING_PLATE
})

# 데이터프레임 출력
df
```
